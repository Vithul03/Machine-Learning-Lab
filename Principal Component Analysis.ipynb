```python
import pandas as pd 
import numpy as np 
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data=pd.read_csv("PCA_CustomerSegmentation.csv") 
print("the First 5 values of data is :\n") 
data.head()
```

```python
Features=['Age','Annual Income','Spending Score','Number of Purchases'] 
X=data[Features] 
```

```python
Scaler=StandardScaler() 
X_scaled=Scaler.fit_transform(X)
```

```python
pca=PCA(n_components=2) 
X_pca=pca.fit_transform(X_scaled)
```

```python
data_pca=pd.DataFrame(data=X_pca,columns=['pc1','pc2']) 
data_pca['Cluster']=data['Cluster'] 
```

```python
plt.figure(figsize=(10,6)) 
sns.scatterplot(x='pc1',y='pc2',hue='Cluster',data=data_pca,palette='viridis') 
plt.title('PCA Result - Customer Segmentation') 
plt.xlabel('principal component 1') 
plt.ylabel('principal component 2') 
plt.show() 
```

```python
explained_variance=pca.explained_variance_ratio_ 
print(f'Explained Variance by each Component: {explained_variance}') 
print(f'Total Explained Variance : {np.sum(explained_variance)}') 
```
